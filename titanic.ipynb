```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
cols = ['class', 'survived', 'sex', 'age', 'sibsp', 'fare', 'embarked']
df = pd.read_csv('data/Titanic Dataset.csv').drop('name', axis=1)
df.columns = cols
```

```python
df['sex'] = (df['sex'] == 'male').astype(int)
df['embarked'] = (df['embarked'] == 'S').astype(int)
```

```python
median_val = df[cols].median()
df[cols] = df[cols].fillna(median_val)
```

```python
for label in cols[:-1]:
        plt.hist(df[df['survived']== 1][label], color= 'blue', label= 'Survived', alpha= 0.7, density= True)
        plt.hist(df[df['survived']== 0][label], color= 'red', label= 'Died', alpha= 0.7, density= True)
        plt.title(label)
        plt.xlabel(label)
        plt.ylabel('Probability')
        plt.legend()
        plt.show()
```

```python
train, test = np.split(df.sample(frac=1), [int(0.6*len(df))])
```

```python
y_train, x_train = train.pop('survived').values, train.values
y_test, x_test = test.pop('survived').values, test.values
```

```python
log_model = LogisticRegression()
log_model = log_model.fit(x_train, y_train)
```

```python
y_pred = log_model.predict(x_test)
print(classification_report(y_test, y_pred))
```
